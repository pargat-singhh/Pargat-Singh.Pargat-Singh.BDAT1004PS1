```
//Question 4
using System;
public class EvenOdd
{
    public static void Main()
    {
        int[] main = new int[10];
        int[] even = new int[10];
        int[] odd = new int[10];
        int i, j = 0, k = 0, n;

        Console.Write("Input the number of elements to be stored in the array :");
        n = Convert.ToInt32(Console.ReadLine());

        Console.Write("Input {0} elements in the array :\n", n);
        for (i = 0; i < n; i++)
        {
            Console.Write("element - {0} : ", i);
            main[i] = Convert.ToInt32(Console.ReadLine());
        }

        for (i = 0; i < n; i++)
        {
            if (main[i] % 2 == 0)
            {
                even[j] = main[i];
                j++;
            }
            else
            {
                odd[k] = main[i];
                k++;
            }
        }

        Console.Write("\nThe Even elements are : \n");
        for (i = 0; i < j; i++)
        {
            Console.Write("{0} ", even[i]);
        }

        Console.Write("\nThe Odd elements are :\n");
        for (i = 0; i < k; i++)
        {
            Console.Write("{0} ", odd[i]);
        }
        Console.Write("\n\n");
    }
}



//References:
// C# - Separate odd and even integers in separate arrays. (2022, August 19). W3resource. 
// Retrieved October 14, 2022, 
// from https://www.w3resource.com/csharp-exercises/array/csharp-array-exercise-10.php
```

```
//Question 5

//Part b

using System;

class GFG
{

    static bool inside(double x1, double y1, double x2,
                          double y2, double x, double y)
    {
        if (x > x1 && x < x2 &&
            y > y1 && y < y2)
            return true;
        else
            return false;
    }

    public static void Main()
    {

        // lower-left and upper-right
        // corners of rectangle
        double x1 = 0.3, y1 = 0.5, 
               x2 = 1.1, y2 = 0.7;     //Change value of x1,y1,x2 and y2 for second case and it gives a "True" result

        double x = 1, y = 1;

        Console.WriteLine("inside({0},{1},{2},{3},{4},{5})", x,y,x1,y1,x2,y2);

        if (inside(x,y,x1,y1,x2,y2))
            Console.Write("True, point lies inside");
        else
            Console.Write("False, point does not lie inside");
    }
}
```
